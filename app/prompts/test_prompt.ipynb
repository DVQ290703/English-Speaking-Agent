```python
import sys
from pathlib import Path

# Walk up from cwd until the app/ directory is found (project root)
project_root = Path.cwd()
for _ in range(5):
    if (project_root / "app").exists():
        break
    project_root = project_root.parent

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from app.prompts.prompt_builder import _load_topics, build_system_prompt

print(f"Project root: {project_root}")
```

```python
topics = _load_topics()
topic_keys = list(topics.keys())

print("Available topics:\n")
for i, key in enumerate(topic_keys, 1):
    print(f"  {i}. {key}")
```

```python
choice = int(input("Enter topic number: ").strip())
chosen_topic = topic_keys[choice - 1]
topic_data = topics[chosen_topic]
sub_keys = list(topic_data["options"].keys())

print(f"\nChose: {chosen_topic}")
print("\nAvailable sub-topics:\n")
for i, key in enumerate(sub_keys, 1):
    print(f"  {i}. {key}")
```

```python
choice = int(input("Enter sub-topic number: ").strip())
chosen_sub = sub_keys[choice - 1]

prompt = build_system_prompt(topic=chosen_topic, sub_option=chosen_sub)

print(f"\nChose: {chosen_sub}")
print("\n" + "=" * 60)
print("FULL SYSTEM PROMPT")
print("=" * 60 + "\n")
print(prompt)
```
